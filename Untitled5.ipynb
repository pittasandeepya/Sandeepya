# HEATMAP(WITHOUT NAN VALUES)

```python
def null_gender(cols):
   gender=cols[0]
   lunch=cols[1]

   if pd.isnull(gender):
     if lunch==1:
         return 48
     elif lunch==2:
         return 39
     else:
           return 44
   else:
        return   gender  

   
```

```python
log['gender']=log[['gender','lunch']].apply(null_gender,axis=1)
```

```python
sns.heatmap(log.isnull())
```

# CLEANING DATASET

```python
log.drop('lunch',axis=1,inplace=True)
```
